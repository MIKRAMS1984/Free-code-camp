# Tips dataset

```python
df = sns.load_dataset('tips')
df.head()
```

# Finding avg. bill using sns.barplot

```python
sns.barplot(x = 'day', y = 'total_bill', data=df);
```

```python
# to get avg.bill for every day.
avg_df = df.groupby('day')[['total_bill']].mean()
avg_df
```

```python
plt.bar(avg_df.index, avg_df.total_bill);
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = df);
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = df);
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'smoker', data = df);
```
